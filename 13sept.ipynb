```python
def city_count(a):
   mask=data["city"].str.lower()==a.lower()
   return data[mask].shape[0]
city_count('Hyderabad')
```

```python
def city_count(a):
   mask0=data["city"].str.lower()==a.lower()
   mask1=data["dl_applied"]==1
   mask=mask0 & mask1
   return data[mask]
city_count('Hyderabad')
```
