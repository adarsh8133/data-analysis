```python
def run_count(batsman_name):
    mask=data['batsman'].str.lower()== batsman_name.lower()
    return data[mask].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
run_count('ms dhoni')
```
